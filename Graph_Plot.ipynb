```python
#importing library
import pandas as pd
from sklearn import tree
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = ['DT1','DT2','DT3','DT4','DT5']
y = [1,1,1,1,1]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('Accuracy Without K-Fold Cross Validation',fontsize = 20)
ax.set_xlabel('Algorithm',fontsize = 20)
ax.set_ylabel('Accuracy(%)',fontsize = 20)
plt.plot(x,y,':ko',label='Accuracy')

ax.legend()
```

```python
x = ['DT1','DT2','DT3','DT4','DT5']
y = [1,1,.9,.9,1]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('Accuracy With K-Fold Cross Validation',fontsize = 20)
ax.set_xlabel('Algorithm',fontsize = 20)
ax.set_ylabel('Accuracy(%)',fontsize = 20)
plt.plot(x,y,'-.r',label='Accuracy')

ax.legend()
```

```python
x = ['DT1','DT2','DT3','DT4','DT5']
y = [1,1,1,1,1]
y2 = [1,1,.9,.9,1]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('Accuracy Graph',fontsize = 20)
ax.set_xlabel('Algorithm',fontsize = 20)
ax.set_ylabel('Accuracy (%)',fontsize = 20)
plt.plot(x,y,':ko',label='Without Cross validation')
plt.plot(x,y2,'-..g',label='With Cross Validation')

ax.legend()
```
